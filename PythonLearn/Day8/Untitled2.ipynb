```python
import pandas as pd

import matplotlib.pyplot as plt

# pip install matplotlib
```

```python
df = pd.read_csv("userfoods.csv")
```

```python
df
```

```python
df2= df.groupby("Item").mean()


df2=df2.reset_index()

df2
```

```python
g1 = plt.figure(figsize=(6,3))

plt.bar(x=df2["Item"], height=df2["Spend"], color=["red","blue"])
plt.ylabel("Spend")
plt.xlabel("Items")
plt.xticks(rotation=40)
plt.title("Items for each price")

plt.show()
plt.savefig("bar.png")

# plt.ylim(0,20)
# plt.yticks(0,10)
```

```python
plt.pie(df2["Spend"], labels=df2["Item"].values,
       autopct='%.2f',textprops={"fontweight":"bold"})


plt.legend(loc="upper left")
plt.show()
```
