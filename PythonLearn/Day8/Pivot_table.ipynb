```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'First Name': ['Aryan', 'Rohan', 'Riya', 'Yash', 'Siddhant', ],
                   'Last Name': ['Singh', 'Agarwal', 'Shah', 'Bhatia', 'Khanna'],
                   'Type': ['Full-time Employee', 'Intern', 'Full-time Employee', 
                            'Part-time Employee', 'Full-time Employee'],
                   'Department': ['Administration', 'Technical', 'Administration', 
                                  'Technical', 'Management'],
                   'YoE': [2, 3, 5, 7, 6],
                   'Salary': [20000, 5000, 10000, 10000, 20000]})
```

```python
df
```

```python
pd.pivot_table(data=df , index="Department",columns=["First Name"] ,values="Salary")
```

```python
df
```

```python
pd.pivot_table(data=df, index=['Type'],
                        values='Salary',
                        aggfunc=["sum","mean"])
```

```python
pd.pivot_table(data=df, index=['Type',"Department"],
                        values='Salary',
                        aggfunc=["sum","mean"])
```
