```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df["new"] = df["Petrol"]+1
df.head(3)
```

```python
colors = ["Purple","Yellow","Green","blue"]
labels=["Diesel","Petrol","new","Gas"]
plt.figure(figsize=(10,6))
plt.stackplot(df["Year"],
             df["Diesel"],
              df["Petrol"],
              df["new"],
              df["Gas"],
              colors=colors,
              edgecolor=None
             )

plt.legend(loc='upper right',labels=labels)
plt.title("Year on Year Vehicle data")
plt.xticks(df["Year"],rotation=90)

plt.show()
```
